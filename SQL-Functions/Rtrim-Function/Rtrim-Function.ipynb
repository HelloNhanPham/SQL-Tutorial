# <span style='display: flex; justify-content: center; color: #E54646'><b>RTRIM() Function</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p>SQL RTRIM function is a string function that removes spaces from the right side of a character or string. RTRIM or Right Trim function is used in data cleaning and manipulation in SQL.</p>

<p>In this example, we will learn the basics of the RTRIM function, and learn how it works with examples of different use cases.</p>

## <span><b>RTRIM() Function in SQL</b></span>

<p>SQL RTRIM() function allows us to remove the trailing empty spaces from the string. <strong>Trailing white spaces</strong> are the redundant white spaces present at the right end of the string. For example, In “GFG “, there are 5 empty white spaces at the right end of the string.</p>

<p>If there are redundant white spaces at the beginning of the string. <strong>For example</strong>: ” GFG”, there are 5 whitespaces at the beginning of the string”, these are called <strong>leading whitespace</strong>.</p>

<p>RTRIM function works in: MySQL 4.0, SQL Server (starting with 2008), Azure SQL Database, Azure SQL Data Warehouse, Parallel Data Warehouse, Oracle, and all other major database systems.</p>

<p><strong>Syntax</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <td style='background-color: #F0F0F0'>
      <strong>
        <span style='color: #BD6B09'>RTRIM</span>(<span style='color: #000000'>input_text</span>);
      </strong>
    </td>
  </tr>
</table>

<p>Or</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>

</table>

## <span><b>SQL RTRIM Function Examples</b></span>

<p>Let’s see some examples of RTRIM function in SQL, and understand it’s working with examples of different use cases.</p>

### <span style='color: #BD6B09'><strong>Example 1: Using RTRIM Function to Remove Trailing Spaces from a String</strong></span>

<p>The RTRIM function in SQL is used to remove trailing spaces from a string. Here are a few examples to illustrate its usage:</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <td style='background-color: #F0F0F0'>
      <strong>
        <span style='color: #BD6B09'>SELECT</span> <span style='color: #000000'>(</span><span style='color: #000000'>'Geeks for Geeks'</span> <span style='color: #BD6B09'>AS</span> <span style='color: #000000'>"Before RTRIM()"</span>, 
        <br><span style='color: #BD6B09'>RTRIM</span>(<span style='color: #000000'>'Geeks for Geeks'</span>) <span style='color: #BD6B09'>AS</span> <span style='color: #000000'>"After RTRIM()"</span>;
      </strong>
    </td>
  </tr>
</table>

<p><strong>Output</strong></p>

<p>As we can see, the whitespaces in the right end of the string are removed, and the left side spaces remain as it is.</p>

### <span style='color: #BD6B09'><strong>Example 3: Using RTRIM Function With a Variable</strong></span>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <td style='background-color: #F0F0F0'>
      <strong>

        <span style='color: #BD6B09'>BEGIN</span><br>
        &nbsp;&nbsp;&nbsp;&nbsp;<span style='color: #BD6B09'>--</span> <span style='color: #000000'>Declare a variable</span><br>
        &nbsp;&nbsp;&nbsp;&nbsp;<span style='color: #BD6B09'>DECLARE</span> <span style='color: #000000'>input_string VARCHAR(15);</span><br>
        &nbsp;&nbsp;&nbsp;&nbsp;<span style='color: #BD6B09'>--</span> <span style='color: #000000'>Assign a value to the variable</span><br>

        &nbsp;&nbsp;&nbsp;&nbsp;<span style='color: #BD6B09'>--</span> <span style='color: #000000'>Use the variable in a query</span><br>
        &nbsp;&nbsp;&nbsp;&nbsp;<span style='color: #BD6B09'>SELECT</span> <span style='color: #BD6B09'>CONCAT</span>(<span style='color: #BD6B09'>RTRIM</span>(<span style='color: #000000'>input_string</span>), <span style='color: #000000'>' World'</span>) <span style='color: #BD6B09'>AS</span> <span style='color: #000000'>result;</span><br>

<p><strong>Output</strong></p>

<p>We declare a variable in <strong>stored procedure</strong> and assign it a value. We changed delimiter to “//” to consider procedure as one statement. Then we use CONCAT function to concatenate declared variable and “World”.We have also use RTRIM function on declare variable in CONCAT function and As we can see in results, RTRIM method removed whitespace s from the declared variable.</p>

## <span><b>Key Takeaways About SQL RTRIM Function</b></span>

<li style='text-indent: 1cm'>RTRIM() function removes the trailing spaces in the given string or removes trailing spaces for all the values present in a specific column in the table.</li>
<li style='text-indent: 1cm'>It allows only one parameter i.e. either a string or a column having values of string type.</li>
<li style='text-indent: 1cm'>It is supported in all major database systems.</li>
<li style='text-indent: 1cm'>When used in <strong>SELECT statements</strong>, the function improves the readability of results by eliminating unnecessary white spaces.</li>
