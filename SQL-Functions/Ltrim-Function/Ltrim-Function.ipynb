# <span style='display: flex; justify-content: center; color: #E54646'><b>LTRIM() Function</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p><strong>The SQL LTRIM function</strong> removes the leading space from the left of a string or string expression.</p>

## <span><b>LTRIM() Function in SQL</b></span>

<p>The LTRIM() function in SQL is an inbuilt string function that is used to remove the <strong>leading empty spaces</strong> from the left side of our data. The LTRIM function can also be used to remove a specific character or substring from the left side of the string.

<p>The <strong>LTRIM function</strong> is primarily used for data cleaning purposes where the data administrator uses this function to remove unnecessary characters from the database in one go.</p>

<p><strong>Syntax</strong></p>

<p>SQL LTRIM function syntax is:</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <td style='background-color: #F0F0F0'>
      <strong>
        <span style='color: #BD6B09'>LTRIM</span>(<span style='color: #BD6B09'>Input_String</span>, [<span style='color: #008000'>Trim_Characters</span>])
      </strong>
    </td>
  </tr>
</table>

<p><strong>Parameters</strong></p>

<li style='text-indent: 1cm'><strong>Input_String</strong>: String from which you want to remove leading spaces.</li>
<li style='text-indent: 1cm'><strong>Trim_Characters</strong>: [optional] specified characters you want to remove.</li>

<p><strong>Note</strong>: If we don’t specify any characters to remove, the LTRIM function removes the white spaces for <strong>data cleaning</strong> and <strong>manipulation</strong> in SQL.</p>

## <span><b>SQL LTRIM function Working</b></span>

<p>The following diagram shows how the <strong>SQL LTRIM function</strong> affects strings in the table.</p>

<div align='center'><img src='https://media.geeksforgeeks.org/wp-content/uploads/20231231005307/LTRIM_Workflow.png' style='border: solid black 5px'><br><small><i>Workflow of LTRIM function</i></small></div>

<p>The function LTRIM in SQL analyzes the input string from the left side and removes any empty spaces or the mentioned character found on that string. The function does not alter the original string, it returns a new string where the leading whitespaces are removed.</p>

<p>This function doesn’t deal with whitespaces present between the strings. For example if the string ‘ JOHN’ was written as ‘ JO HN’ then only the leading white spaces on the left side will be removed by this function and the spaces in between will remain as they were earlier.</p>

## <span><b>Important Points About SQL LTRIM Function</b></span>

<li style='text-indent: 1cm'><strong>SQL LTRIM</strong> is useful inbuilt function which can be used to remove the whitespaces on left our data in the databases.</li>
<li style='text-indent: 1cm'>Not only whitespaces it can also remove a specific character or a subtring from the data.</li>
<li style='text-indent: 1cm'>LTRIM function in SQL can be used on entire column/field of table to remove whitespace.</li>
<li style='text-indent: 1cm'>LTRIM function only works for the left side of data, and will not affect the middle or right side of data.</li>
