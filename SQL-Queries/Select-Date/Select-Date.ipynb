# <span style='display: flex; justify-content: center; color: #E54646'><b>SQL SELECT DATE</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p>In Microsoft SQL Server, <strong>SELECT DATE</strong> is used to get the data from the table related to the date, the default format of date is ‘YYYY-MM-DD’.</p>

<strong>Syntax</strong>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>SELECT</span> * <span style='color: #BD6B09'>FROM</span> table_name<br><span style='color: #BD6B09'>WHERE</span> condition1, condition2,..;</pre></strong></th>
    </tr>
</table>
