# <span style='display: flex; justify-content: center; color: #E54646'><b>WHAT IS TEMPORARY TABLE IN SQL?</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p><b>Temporary Tables</b> are most likely as Permanent Tables. Temporary Tables are Created in TempDB and are automatically deleted as soon as the last connection is terminated. Temporary Tables helps us to store and process intermediate results. Temporary tables are very useful when we need to store temporary data. The Syntax to create a Temporary Table is given below:</p>

### <span style='color: #BD6B09'><strong>To Create Temporary Table</strong></span>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>CREATE TABLE</span> #EmpDetails (id INT, name VARCHAR(25));</pre></strong></th>
    </tr>
</table>

### <span style='color: #BD6B09'><strong>To Insert Values Into Temporary Table</strong></span>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>INSERT INTO</span> #EmpDetails <span style='color: #BD6B09'>VALUES</span> (01, 'Lalit'), (02, 'Atharva')</pre></strong></th>
    </tr>
</table>

### <span style='color: #BD6B09'><strong>To Select Values From Temporary Table</strong></span>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>SELECT</span> * <span style='color: #BD6B09'>FROM</span> #EmpDetails </pre></strong></th>
    </tr>
</table>

### <span style='color: #BD6B09'><strong>Result</strong></span>

<table align='center'>
    <tr align = 'center'>
        <th><strong>id</strong></th>
        <th><strong>name</strong></th>
    </tr>
    <tr align='center'>
        <td>1</td>
        <td>Lalit</td>
    </tr>
    <tr align='center'>
        <td>2</td>
        <td>Atharva</td>
    </tr>
</table>

<p>There are 2 types of Temporary Tables: Local Temporary Table, and Global Temporary Table. These are explained as following below.</p>

### <span style='color: #BD6B09'><strong>1. Local Temporary Table</strong></span>

<p>A Local Temp Table is available only for the session that has created it. It is automatically dropped (deleted) when the connection that has created it, is closed. To create Local Temporary Table Single “#” is used as the prefix of a table name.</p>

<p>Also, the user can drop this temporary table by using the “DROP TABLE #EmpDetails” query. There will be Random Numbers are appended to the Name of Table Name. If the Temporary Table is created inside the stored procedure, it get dropped automatically upon the completion of stored procedure execution.</p>

### <span style='color: #BD6B09'><strong>Example</strong></span>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>CREATE PROCEDURE</span> ProcTemp<br><span style='color: #BD6B09'>AS</span><br><span style='color: #BD6B09'>BEGIN</span><br><span style='color: #BD6B09'>CREATE TABLE</span> #EmpDetails<br><span style='color: #BD6B09'>INSERT INTO</span> #EmpDetails <span style='color: #BD6B09'>VALUES</span> ( 01, 'Lalit'), ( 02, 'Atharva')<br><span style='color: #BD6B09'>SELECT</span> * <span style='color: #BD6B09'>FROM</span> #EmpDetails<br><span style='color: #BD6B09'>END<br>EXECUTE</span> ProcTemp</pre></strong></th>
    </tr>
</table>

### <span style='color: #BD6B09'><strong>2. Global Temporary Table</strong></span>

<p>To create a Global Temporary Table, add the “##” symbol before the table name.</p>

<table align='center'>
    <tr style='background-color: #F0F0F0; color: black; border: solid white'>
        <th style='padding-top: 20px; padding-bottom: 0px;'><strong><pre><span style='color: #BD6B09'>CREATE TABLE</span> ##EmpDetails (id INT, name VARCHAR(25))</pre></strong></th>
    </tr>
</table>

<p>Global Temporary Tables are visible to all connections and Dropped when the last connection referencing the table is closed. Global Table Name must have an Unique Table Name. There will be no random Numbers suffixed at the end of the Table Name.</p>
