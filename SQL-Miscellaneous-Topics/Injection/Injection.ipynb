# <span style='display: flex; justify-content: center; color: #E54646'><b>Injection</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p><strong>SQL injection</strong> is a code injection technique attackers use to gain unauthorized access to a database by injecting malicious SQL commands into web page inputs.</p>

<p>Attackers can extract sensitive information, modify database data, execute administration operations on the database (such as shutdown DBMS), recover the content of a given file present on the DBMS file system, and in some cases, issue commands to the operating system.</p>

<p>In this article, we will discuss what is SQLi(SQL Injection), Types of SQL injection, SQL injection in web pages, how to prevent SQL injection attacks, and many more.</p>

#### <span style='color: #A0522D'><strong>What is SQL Injection?</strong></span>

<p><strong>SQLi</strong> or <strong>SQL Injection</strong> is a web page vulnerability that lets an attacker make queries with the database. Attackers take advantage of web application vulnerability and inject an SQL command via the input from users to the application.</p>

<p>Attackers can SQL queries like SELECT to retrieve confidential information which otherwise wouldn’t be visible. SQL injection also lets the attacker to perform a <strong>denial-of-service (DoS) attacks</strong> by overloading the server requests.</p>

#### <span style='color: #A0522D'><strong>The Exploitation of SQL Injection in Web Applications</strong></span>

<p>Web servers communicate with database servers anytime they need to retrieve or store user data. SQL statements by the attacker are designed so that they can be executed while the web server is fetching content from the application server.</p>

<p><strong>SQL in Web Pages</strong></p>

<p>SQL injection typically occurs when you ask a user for input, such as their username/user ID, and instead of their name/ID, the user inputs an SQL statement that will be executed without the knowledge about your database.</p>

<p>For example,</p>

#### <span style='color: #A0522D'><strong>SQL Injection Types</strong></span>

<p>There are different types of SQL injection attacks:</p>

<p><strong>1. In-band SQL Injection</strong></p>

<p>It involves sending malicious SQL queries directly through the web application’s interface and allows attackers to extract sensitive information or modify the database itself.</p>

<p><strong>2. Error-based SQL Injection</strong></p>

<p>Attackers exploit error messages generated by the web application by analyzing error messages to gain access to confidential data or modify the database.</p>

<p><strong>3. Blind SQL Injection</strong></p>

<p>Attackers send malicious SQL queries and observe the application’s response. By analyzing the application’s behavior, attackers can determine the success of the query.</p>

<p><strong>4. Out-of-band SQL Injection</strong></p>

<p>Uses a different channel to communicate with the database. Allows attackers to exfiltrate sensitive data from the database.</p>

<p><strong>5. Inference-based SQL Injection</strong></p>

<p>Uses statistical inference to gain access to confidential data. Attackers create queries that return the same result regardless of input values.</p>

#### <span style='color: #A0522D'><strong>Impact of SQL Injection</strong></span>

<p>The hacker can retrieve all the user data present in the database such as user details, credit card information, and social security numbers, and can also gain access to protected areas like the administrator portal. It is also possible to delete user data from the tables.</p>

<p>Nowadays, all online shopping applications and bank transactions use back-end database servers. So in case the hacker is able to exploit SQL injection, the entire server is compromised.</p>

#### <span style='color: #A0522D'><strong>SQL Injection Prevention</strong></span>

<p>Developers can use the following prevention measures to prevent SQL injection attacks.</p>

<table style='background-color: #FCD9C4'>
  <tr>
    <td style='border: solid white'>
      <li><strong>User Authentication</strong> − Validating input from the user by pre-defining the length, type of input, and authenticating the user.</li>
      <li><strong>Restricting access privileges</strong> − Defining how much data any outsider can access from the database. Users should not be granted permission to access everything in the database.</li>
      <li><strong>Do not use system administrator accounts</strong> − Avoid using system administrator accounts for routine tasks to minimize security risks.</li>
    </td>
  </tr>
</table>
