## <span style='color: #EE7C6B'><b>Short Overview of SQL</b></span>

<div align='justify'>
    <p><b>Structured Query Language (SQL)</b>, as we all know, is the database language by which we can perform certain operations on the existing database, and we can also use this language to create a database. SQL uses certain commands like CREATE, DROP, INSERT, etc. to carry out the required tasks. </p>
    <p>SQL commands are like instructions to a table. It is used to interact with the database with some operations. It is also used to perform specific tasks, functions, and queries of data. SQL can perform various tasks like creating a table, adding data to tables, dropping the table, modifying the table, set permission for users.</p>
    <p>These SQL commands are mainly categorized into five categories: </p>
</div>

1. <b>DDL</b> – Data Definition Language
2. <b>DQL</b> – Data Query Language
3. <b>DML</b> – Data Manipulation Language
4. <b>DCL</b> – Data Control Language
5. <b>TCL</b> – Transaction Control Language

<div align='justify'>
    <p>Now, we will see all of these in detail.</p>
</div>

## <span style='color: #EE7C6B'><b>Important SQL Commands</b></span>

<div align='justify'>
    <p>Some of the most important SQL commands are:</p>

1. <b>SELECT</b>: Used to retrieve data from a database.
2. <b>INSERT</b>: Used to add new data to a database.
3. <b>UPDATE</b>: Used to modify existing data in a database.
4. <b>DELETE</b>: Used to remove data from a database.
5. <b>CREATE TABLE</b>: Used to create a new table in a database.
6. <b>ALTER TABLE</b>: Used to modify the structure of an existing table.
7. <b>DROP TABLE</b>: Used to delete an entire table from a database.
8. <b>WHERE</b>: Used to filter rows based on a specified condition.
9. <b>ORDER BY</b>: Used to sort the result set in ascending or descending order.
10. <b>JOIN</b>: Used to combine rows from two or more tables based on a related column between them.
</div>
