# <span style='display: flex; justify-content: center; color: #E54646'><b>EXISTS</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p>The EXISTS condition in SQL is used to check whether the result of a correlated nested query is empty (contains no tuples) or not. The result of EXISTS is a boolean value True or False. It can be used in a SELECT, UPDATE, INSERT or DELETE statement.</p>

<p><strong>Syntax</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT</span> column_name(s) ><span style='color: #BD6B09'>FROM</span> table_name ><span style='color: #BD6B09'>WHERE EXISTS</span> (><span style='color: #BD6B09'>SELECT</span> column_name(s) ><span style='color: #BD6B09'>FROM</span> table_name ><span style='color: #BD6B09'>WHERE</span> condition);</pre></strong></th></tr>
</table>

### <span style='color: #BD6B09'><strong>Queries</strong></span>

<p><strong>1. Using EXISTS condition with SELECT statement</strong></p> 
<p>To fetch the first and last name of the customers who placed atleast one order.</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT</span> fname, lname 
<span style='color: #BD6B09'>FROM</span> Customers 
<span style='color: #BD6B09'>WHERE EXISTS</span> (
    <span style='color: #BD6B09'>SELECT</span> * 
    <span style='color: #BD6B09'>FROM</span> Orders 
    <span style='color: #BD6B09'>WHERE</span> Customers.customer_id = Orders.c_id
);
</pre></strong></th></tr>
</table>

<p><strong>2. Using NOT with EXISTS</strong></p>

<p>Fetch last and first name of the customers who has not placed any order.</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT</span> lname, fname
<span style='color: #BD6B09'>FROM</span> Customers
<span style='color: #BD6B09'>WHERE NOT EXISTS</span> (
    <span style='color: #BD6B09'>SELECT</span> * 
    <span style='color: #BD6B09'>FROM</span> Orders 
    <span style='color: #BD6B09'>WHERE</span> Customers.customer_id = Orders.c_id
);</pre></strong></th></tr>
</table>

<p><strong>3. Using EXISTS condition with DELETE statement</strong></p>

<p>Delete the record of all the customer from Order Table whose last name is ‘Mehra’.</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>DELETE</span> 
<span style='color: #BD6B09'>FROM</span> Orders
<span style='color: #BD6B09'>WHERE EXISTS</span> (
    <span style='color: #BD6B09'>SELECT</span> *
    <span style='color: #BD6B09'>FROM</span> Customers
    <span style='color: #BD6B09'>WHERE</span> Customers.customer_id = Orders.c_id
    <span style='color: #BD6B09'>AND</span> Customers.lname = 'Mehra'
);
<span style='color: #BD6B09'>SELECT</span> * <span style='color: #BD6B09'>FROM</span> Orders;
</pre></strong></th></tr>
</table>

<p><strong>4. Using EXISTS condition with UPDATE statement</strong></p>

<p>Update the lname as ‘Kumari’ of customer in Customer Table whose customer_id is 401.</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>UPDATE</span> Customers
<span style='color: #BD6B09'>SET</span> lname = 'Kumari'
<span style='color: #BD6B09'>WHERE EXISTS</span> (
    <span style='color: #BD6B09'>SELECT</span> *
    <span style='color: #BD6B09'>FROM</span> Customers
    <span style='color: #BD6B09'>WHERE</span> customer_id = 401
);
<span style='color: #BD6B09'>SELECT</span> * <span style='color: #BD6B09'>FROM</span> Customers;
</pre></strong></th></tr>
</table>
