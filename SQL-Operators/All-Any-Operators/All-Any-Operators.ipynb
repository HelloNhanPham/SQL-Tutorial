# <span style='display: flex; justify-content: center; color: #E54646'><b>ALL, ANY OPERATOR</b></span>

<div align='justify'>
<blockquote style='background-color: white; color: black'>

<p><strong>ALL & ANY</strong> are logical operators in SQL. They return boolean value as a result.</p>

## <span style='display: flex; justify-content: center; color: #F19373'><strong>ALL OPERATOR</strong></span>

<p>ALL operator is used to select all tuples of SELECT STATEMENT. It is also used to compare a value to every value in another value set or result from a subquery.</p>

<li style='text-indent: 1cm'>The ALL operator returns TRUE if all of the subqueries values meet the condition. The ALL must be preceded by comparison operators and evaluates true if all of the subqueries values meet the condition.</li>
<li style='text-indent: 1cm'>ALL is used with SELECT, WHERE, HAVING statement.</li>

### <span style='color: #BD6B09'><strong>ALL with SELECT Statement</strong></span>

<p><strong>Syntax</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT ALL</span> field_name
<span style='color: #BD6B09'>FROM</span> table_name
<span style='color: #BD6B09'>WHERE</span> condition(s);</pre></strong></th></tr>
</table>

### <span style='color: #BD6B09'><strong>ALL with WHERE or HAVING Statement</strong></span>

<p><strong>Syntax</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT</span> column_name(s) <span style='color: #BD6B09'>FROM</span> table_name <span style='color: #BD6B09'>WHERE</span> column_name comparison_operator <span style='color: #BD6B09'>ALL</span>
(<span style='color: #BD6B09'>SELECT</span> column_name <span style='color: #BD6B09'>FROM</span> table_name <span style='color: #BD6B09'>WHERE</span> condition(s));
</pre></strong></th></tr>
</table>

## <span style='display: flex; justify-content: center; color: #F19373'><strong>ANY OPERATOR</strong></span>

<p>ANY compares a value to each value in a list or results from a query and evaluates to true if the result of an inner query contains at least one row.</p>

<li style='text-indent: 1cm'>ANY return true if any of the subqueries values meet the condition.</li>
<li style='text-indent: 1cm'>ANY must be preceded by comparison operators.</li>

<p><strong>Syntax</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT</span> column_name(s) <span style='color: #BD6B09'>FROM</span> table_name <span style='color: #BD6B09'>WHERE</span> column_name comparison_operator <span style='color: #BD6B09'>ANY</span>
(<span style='color: #BD6B09'>SELECT</span> column_name <span style='color: #BD6B09'>FROM</span> table_name <span style='color: #BD6B09'>WHERE</span> condition(s));</pre></strong></th></tr>
</table>

<p><strong><span style='color: red'>Query 1</span> - Find the Distinct CategoryID of the products which have any record in OrderDetails Table.</strong></p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
<tr><th style='background-color: #F0F0F0'><strong><pre><span style='color: #BD6B09'>SELECT DISTINCT</span> CategoryID <span style='color: #BD6B09'>FROM</span> Products 
<span style='color: #BD6B09'>WHERE</span> ProductID = <span style='color: #BD6B09'>ANY</span> (<span style='color: #BD6B09'>SELECT</span> ProductID <span style='color: #BD6B09'>FROM</span> OrderDetails);</pre></strong></th></tr>
</table>
