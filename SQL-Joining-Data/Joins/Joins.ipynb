<p><strong>SQL Join</strong></p> operation combines data or rows from two or more tables based on a common field between them.

<p>In this article, we will learn about <strong>Joins in SQL</strong>, covering JOIN types, syntax, and examples.</p>

## <span style='color: '><strong>SQL JOIN</strong></span>

<p>SQL JOIN clause is used to query and access data from multiple tables by establishing logical relationships between them. It can access data from multiple tables simultaneously using common key values shared across different tables.</p>

<p>We can use SQL JOIN with multiple tables. It can also be paired with other clauses, the most popular use will be using JOIN with <strong>WHERE</strong> clause to filter data retrieval.</p>

## <span style='color: '><strong>Types of JOIN in SQL</strong></span>

<p>There are many types of Joins in SQL. Depending on the use case, you can use different type of SQL JOIN clause.</p>

<p>Here are the frequently used SQL JOIN types:</p>

<li style='text-indent: 1cm'>INNER JOIN</li>
<li style='text-indent: 1cm'>LEFT JOIN</li>
<li style='text-indent: 1cm'>RIGHT JOIN</li>
<li style='text-indent: 1cm'>FULL JOIN</li>
<li style='text-indent: 1cm'>NATURAL JOIN</li>

## <span style='color: '><strong>SQL INNER JOIN</strong></span>

<p>The <strong>INNER JOIN</strong> keyword selects all rows from both the tables as long as the condition is satisfied. This keyword will create the result-set by combining all rows from both the tables where the condition satisfies i.e value of the common field will be the same.</p>

<p><strong>Syntax</strong></p>

<p>The syntax for SQL INNER JOIN is:</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <th style='background-color: #F0F0F0'>
      <strong>
        <pre>
<span style='color: #BD6B09'>SELECT</span> table1.column1, table1.column2, table2.column1, ...
<span style='color: #BD6B09'>FROM</span> table1
<span style='color: #BD6B09'>INNER JOIN</span> table2
<span style='color: #BD6B09'>ON</span> table1.matching_column = table2.matching_column;
</pre>
      </strong>
    </th>
  </tr>
</table>

<p>Here,</p>

<li style='text-indent: 1cm'><strong>table1</strong>: First table.</li>
<li style='text-indent: 1cm'><strong>table2</strong>: Second table.</li>
<li style='text-indent: 1cm'><strong>matching_column</strong>: Column common to both the tables.</li>

<p><strong style='color: red'>Note</strong> We can also write JOIN instead of INNER JOIN. JOIN is same as INNER JOIN. </p>

## <span style='color: '><strong>SQL LEFT JOIN</strong></span>

<p>LEFT JOIN returns all the rows of the table on the left side of the join and matches rows for the table on the right side of the join. For the rows for which there is no matching row on the right side, the result-set will contain null. LEFT JOIN is also known as LEFT OUTER JOIN.</p>

<p><strong>Syntax</strong></p>

<p>The syntax of LEFT JOIN in SQL is:</p>

<table align='center' style='border: solid white; padding-top: 20px; padding-bottom: 0px;'>
  <tr>
    <th style='background-color: #F0F0F0'>
      <strong>
        <pre>
<span style='color: #BD6B09'>SELECT</span> table1.column1, table1.column2, table2.column1, ...
<span style='color: #BD6B09'>FROM</span> table1
<span style='color: #BD6B09'>LEFT JOIN</span> table2
<span style='color: #BD6B09'>ON</span> table1.matching_column = table2.matching_column;
</pre>
      </strong>
    </th>
  </tr>
</table>

<p>Here,</p>

<li style='text-indent: 1cm'><strong>table1</strong>: First table.</li>
<li style='text-indent: 1cm'><strong>table2</strong>: Second table.</li>
<li style='text-indent: 1cm'><strong>matching_column</strong>: Column common to both the tables.</li>

<p><strong style='color: red'>Note</strong> We can also use LEFT OUTER JOIN instead of LEFT JOIN, both are the same.</p>

## <span style='color: '><strong>SQL Natural join (?)</strong></span>

<p>Natural join can join tables based on the common columns in the tables being joined. A natural join returns all rows by matching values in common columns having same name and data type of columns and that column should be present in both tables.</p>

<p>Both table must have at least one common column with same column name and same data type.</p>

<p>The two table are joined using <strong>Cross join</strong>.</p>

<p>DBMS will look for a common column with same name and data type Tuples having exactly same values in common columns are kept in result.</p>
