<p><strong>SQL CROSS JOIN</strong> returns all the records from the left and right tables. CROSS JOIN returns a combination of each row in the left table paired with each row in the right table.</p>

## <span style='color: '><strong>CROSS JOIN in SQL</strong></span>

<p><strong>Cross Join in SQL</strong> produces a result set that contains the cartesian product of two or more tables. Cross join is also called a <strong>Cartesian Join</strong>.</p>

<p>When CROSS JOIN is used with a <strong>WHERE clause</strong>, it behaves like <strong>INNER JOIN</strong>, filtering the results based on specific conditions.</p>

<p>CROSS JOIN is the best choice when we need to match each row of one table to every other row of another table. It is helpful in many applications where we need to obtain paired combinations of records.</p>

## <span style='color: '><strong>Important Points About CROSS JOIN</strong></span>

<li style='text-indent: 1cm'>CROSS JOIN performs the cross-product of records from two or more joined tables.</li>
<li style='text-indent: 1cm'>It is used when we want every possible combination of rows to be present in a database’s tables.</li>
<li style='text-indent: 1cm'>SQL CROSS JOIN with condition of WHERE Clause operates as an INNER JOIN; when used without one, it produces the cartesian product of all the rows from all the tables provided in the SQL query.</li>
<li style='text-indent: 1cm'>CROSS JOIN is different from other join types like INNER JOIN, <strong>LEFT JOIN</strong>, and <strong>RIGHT JOIN</strong>, as it does not require a matching condition between the tables.</li>
